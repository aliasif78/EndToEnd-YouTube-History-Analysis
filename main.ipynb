```python
# Adding Boolean 'Subcribed' column by comparing 'history' dataFrame with 'subscriptions' dataFrame
tempList = []
channelName = ''

for channelName in history['Channel Name']:
    if channelName.strip() in subscriptions['Name'].str.strip().values:
        tempList.append('Yes')
    else:
        tempList.append('No')

history['Subscribed'] = tempList
history
```

```python
# Rearranging Columns
history = history[['Channel Name', 'Subscribed', 'Video Title', 'Date & Time']]
history
```

```python
# Extract dates from 'Date & Time' column
dates = history['Date & Time'].str.split(' ', n=1, expand=True)[0]

# Calculate value counts for each date
date_counts = dates.value_counts()

# Drop 'count' column
#history.drop(columns='count', inplace=True)

# Add a new column to the DataFrame
history['Videos Watched on this Day'] = dates.map(date_counts)

# Display the updated DataFrame
history
```
